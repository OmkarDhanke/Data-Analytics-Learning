## **Condition Statment If-Else Nested Lader**




```python
p = int(input("Enter Percentage: "))
if p <= 100:
    if p <= 80:
        if p <= 70:
            if p <= 60:
                if p <= 50:
                    if p <= 35:
                        if p <= 0:
                            print("Invalid")
                        else:
                            print("Fail")
                    else:
                        print("D")
                else:
                    print("C")
            else:
                print("B")
        else:
            print("A")
    else:
        print("A+")
else:
    print("Invalid Percentage")
```

```python
a = float(input("Enter Salary: "))
if a > 0:
    if a >= 55000:
        print(f"salary amt = {a} \n bonus value = {a/100*25} \n total salary = {a+(a/100*25)}")
    else:
        if a<= 55000 and a > 40000:
            print(f"salary amt = {a} \n bonus value = {a/100*15} \n total salary = {a+(a/100*15)}")
        else:
            if a<= 40000 and  a > 30000:
                print(f"salary amt = {a} \n bonus value = {a/100*10} \n total salary = {a+(a/100*10)}")
            else:
                if a <= 30000 and a > 20000:
                    print(f"salary amt = {a} \n bonus value = {a/100*7} \n total salary = {a+(a/100*7)}")
                else:
                    if a <= 20000 and a > 8000:
                        print(f"salary amt = {a} \n bonus value = {a/100*5} \n total salary = {a+(a/100*5)}")
                    else:
                        print(f"salary amt = {a} \n bonus value = 0 \n total salary = {a}")
else:
    print("Invalid Salary")
```
