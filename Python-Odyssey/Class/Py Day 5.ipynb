# Inmutable data types **List**


```python
a = [1,2]
print(id(a))
a.append(3)
print(a)
print(id(a))
```

```python
# list in the list with append
a = [1,2]
print(id(a))
# append
a.append(3)
print(a)
a.append([4,5])
print(a)
# Pop
a.pop(3)
print(a)

# extend
a.extend([4,5])
print(a)

a.extend([6,7])
print(a)

a.extend("Ram")
print(a)
```

```python
g = [1,6,1,7,8]
g.remove(1)
print(g)
# remove
g.remove(7,8)

```

```python
# insert
g.insert(3,9)
print(g)
```

```python
# sort
j = [3,4,9,5,7,6,8]
j.sort()
print(j)
```

```python
# reverse
q = [4,6,8,2,5,7,12,14,22]

q.reverse()
print(q)
```

```python
# reverse with sort
h = [4,6,8,2,5,7,12,14,22]
h.sort(reverse=True)
print(h)
```

```python
# Function to calculate area of a triangle
def calculate_triangle_area(base, height):
    area = 0.5 * base * height
    return area

# Input from user
base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))

# Calculate and display the area
area = calculate_triangle_area(base, height)
print(f"The area of the triangle is: {area}")
```
