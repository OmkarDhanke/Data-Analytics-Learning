```python
# OR,and, NOt
a = 20
b = 30
print((a!=b)or(a==b))
print(True or False)
print((a+10)==b)
print (((a+10)==b)or(a!=b))
print(30 and 20)
print(False and (1-1))
print((a!=b) and not(a==b))
```

```python
# Identity is , is Not
a = (1,2)
print((1,2) is a)
print(a is 1)
print(a is b)
```

```python
# Membership in, in not
a = (1,2)
print(2 in a)
print(3 in a)
print((1,2) in a)
print(6 not in a)
print(6 not in 6)
```

```python
x = [1,2,(1,2)]
print(a in x)
print(1 not in a and 2 in x)
print(a in 1)
```

```python
# Assignment +=,-=,/=,//=,*=
a = 20
a += 2
print(a)
a -= 2
print(a)
a /= 2
print(a)
a *= 2
print(a)
a //= 2
print(a)
a %= 2
print(a)
```
