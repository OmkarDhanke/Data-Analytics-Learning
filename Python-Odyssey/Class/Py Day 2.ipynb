```python
# 2 way variable swaping
a = 10
print(a)
print(id(a))

b = 20
print(b)
print(id(b))

a = a+b
print(a)

b = a-b
print(b)

a = a-b
print(a)

print(a)
print(b)
print(id(a))
print(id(b))
```

```python
# 3 way Swaping
a = 100
print(id(a))

c = 'omkar'
print(id(c))

d = a
a = c
c = d

print(a)
print(id(a))
print(c)
print(id(c))
```
