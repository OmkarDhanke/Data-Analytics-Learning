```python
# Write a program to checke given year is leap year or not
a = int(input("Enter the year: "))
if a%4 == 0 and a %100 != 0 or a%400 == 0:
    print(a, "is a leap year")
else:
    print(a, "is not a leap year")
```

```python
# Write a program to gives grade on the percentage if percentage 100 to 80 give A+  70 to 80 A 60 to 70 b 35 to 50 C and 0 to

a = int(input("Enter the percentage: "))
if a<=100 and a>=80:
    print("A+")
elif a>=80 and a<70:
    print("A")
elif a>=70 and a< 60:
    print("B")
elif a>= 60 and a<50:
    print("C")
elif a>=50 and a<35:
    print("D")
elif a>=35 and a<0:
    print("F")
else:
    print("Invalid Marks")
```
