```python
n = int(input("Enter a No:- "))
n1 = n
rev = 0
while n != 0:
  rem = n % 10
  rev = (rev * 10) + rem
  n = n // 10
print(rev)
print(n1)
if n1 == rev:
  print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
n = int(input("Enter a No:- "))
n1 = n
rev = 0
while n != 0:
    for i in range(1, 2):
        print(i, end=" ")
    rem = n % 10
    rev = (rev * 10) + rem
    n = n // 10
print(rev)
print(n1)
if n1 == rev:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
