# **Day 14 While in While Loop**


```python
a = 1
while a <= 4:
  print(chr(64 + a),end = " ")
  a += 1
```

```python
x = 1
while x <= 5:
  a = 1
  while a <= 4:
    print(chr(64 + a),end = " ")
    a += 1
  print()
  x += 1
```

```python
x = 1
while x <= 4:
  a = 1
  while a <= x:
    print("*",end = " ")
    a += 1
  print()
  x += 1
```

```python
x = 1
while x <= 4:
  a = 1
  while a <= x:
    print(a,end = " ")
    a += 1
  print()
  x += 1
```

```python
x = 1
while x <= 4:
  a = 1
  while a <= x:
    print("*",end = " ")
    a += 1
  b = 1
  while b <= 5 - x:
    print( b,end = " ")
    b += 1
  print()
  x += 1
```

```python
x = 5
while x >= 1:
  a = 1
  while a <= x :
    print(a,end = " ")
    a += 1
  print()
  x -= 1
```

# **Function Zip**


```python
# Zip
a = (1,2,3,4,5,6)
b = ('a','b','c','d','e','f')
zip(a,b)
```

```python
# Zip with set
a = (1,2,3,4,5,6)
b = ('a','b','c','d','e','f','g')
set(zip(a,b))
```

```python
# Zip with Dict
a = (1,2,3,4,5,6)
b = ('a','b','c','d','e','f','g')
dict(zip(a,b))
```

```python
# Zip with Dict
x = ('10k','20k','30k')
y = ('raj','karan','sahil')
z = dict(zip(y,x))
print(z)
```
